```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, f1_score, classification_report
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.utils.class_weight import compute_class_weight
from collections import Counter
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE

pd.set_option("display.max_columns", 200)
pd.options.display.float_format = '{:.2f}'.format
```

```python
# data loading
dataset = pd.read_csv('./itb_test_ds_v1.1.csv')
print(dataset.shape)
dataset.head(50)
```

Checking for a correlation matrix fingerprint:

```python
classes = np.unique(dataset.Target)
fig=plt.figure(figsize=(20,100))
columns = 3
rows = 30
i = 1
for c in classes:
    an = dataset[dataset['Target'] == c]
    fig.add_subplot(rows, columns, i)
    cor = an.corr()
    sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
    i = i + 1
    plt.show()
```

```python
target = dataset['Target']
train = dataset.drop('Target', axis=1)
```

```python

label_encoder = LabelEncoder()
label_encoder.fit(classes)
```

```python
label_encoder.classes_
```

```python
target = label_encoder.transform(target)
```

```python
dataset['address'].describe()
```

```python
train_no_address = train.drop(['address','feature01','feature02'], axis=1)
```

#### Train / Validation / Test datasets

```python
X_train, X_test, y_train, y_test = train_test_split(train_no_address, target, test_size=0.2, stratify=target)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.15, random_state=17, stratify=y_train)
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=17)
random_forest.fit(X_train, y_train)
```

```python
rf_predictions = random_forest.predict(X_valid)
```

```python
score = f1_score(y_valid, rf_predictions, average='macro')
print('f1_Score is: ', score, ' for a basic random forest')
```

```python
pd.Series(y_valid).value_counts()
```

```python
matrix = confusion_matrix(y_valid, rf_predictions)
matrix
```

```python
plot_confusion_matrix(
    random_forest, X_valid, y_valid,
    cmap=plt.cm.Blues, display_labels=label_encoder.classes_,
)
plt.title('validation dataset score', color='red')
plt.show()
```

```python
print(classification_report(y_valid, rf_predictions, target_names=label_encoder.classes_))
```

Random forest con balanceo de clases, para balancear en los pesos el enorme imbalance entre los dato

```python
class_weights = compute_class_weight(class_weight='balanced',classes=np.unique(target), y=target)
class_weights = dict(enumerate(class_weights))
class_weights
```

```python
random_forest_balanced = RandomForestClassifier(
    random_state=17,
    class_weight=class_weights,
    warm_start=True
)
random_forest_balanced.fit(X_train, y_train)
rf_balanced_pred = random_forest_balanced.predict(X_valid)
matrix = confusion_matrix(y_valid, rf_balanced_pred)
matrix
```

```python
plot_confusion_matrix(
    random_forest_balanced, train_no_address, target,
    cmap=plt.cm.Blues, display_labels=label_encoder.classes_,
)
plt.title('whole dataset score', color='red')
plt.show()
```

```python
plot_confusion_matrix(
    random_forest_balanced, X_valid, y_valid,
    cmap=plt.cm.Blues, display_labels=label_encoder.classes_,
)
plt.title('validation dataset score', color='red')
plt.show()
```

```python
score = f1_score(y_valid, rf_balanced_pred, average='macro')
print('f1_Score is: ', score, ' for a balanced random forest')
```

```python
print(classification_report(y_valid, rf_balanced_pred, target_names=label_encoder.classes_))
```

Al balanceo del random forest se puede agregar undersample la clase con mas ejemplos ('Class05'). Esto es, reducir la cantidad de ejemplos que se usan para entrenar el modelo:

```python
from imblearn.under_sampling import RandomUnderSampler

under_dict = Counter(y_train)
under_dict[4] = 2000
rus = RandomUnderSampler(random_state=17, sampling_strategy=under_dict)
X_resampled, y_resampled = rus.fit_resample(X_train, y_train)
```

Y al reves, tambien usamos un metodo para crear nuevas instancias en las categorias menos representadas:

```python
oversample = SMOTE(random_state=17)
X_oversampled, y_oversampled = oversample.fit_resample(X_resampled, y_resampled)
oversampled_predictions = random_forest.fit(X_oversampled, y_oversampled).predict(X_valid)
matrix = confusion_matrix(y_valid, oversampled_predictions)
matrix
```

```python
random_forest_pred = random_forest.predict(X_valid)
score = f1_score(y_valid, random_forest_pred, average='macro')
print('f1_Score is: ', score, ' for a basic random forest')
```

```python
print(classification_report(y_valid, random_forest_pred, target_names=label_encoder.classes_))
```

```python
plot_confusion_matrix(
    random_forest_balanced, X_oversampled, y_oversampled,
    cmap=plt.cm.Blues, display_labels=label_encoder.classes_,
)
plt.title('whole dataset confusion matrix', color='red')
plt.show()
```

```python
plot_confusion_matrix(
    random_forest_balanced, X_valid, y_valid,
    cmap=plt.cm.Blues, display_labels=label_encoder.classes_,
)
plt.show()
```

```python
random_forest_balanced_pred = random_forest_balanced.predict(X_valid)
```

```python
score = f1_score(y_valid, random_forest_balanced_pred, average='macro')
print('f1_Score is: ', score, ' for a balanced random forest')
```

```python
print(classification_report(y_valid, random_forest_balanced_pred, target_names=label_encoder.classes_))
```

El caso ultimo parece ser el que mejor resultado da hasta el momento, junto al primer random forest usando datos sin procesar. Esto muestra el alto performance del algoritmo del random forest, ya que necesita poca optimizacion para alcanzar sus mejores resultados.

#### Unsupervised model: KMeans

```python
scaler = StandardScaler()
train_features = scaler.fit(train_no_address)
X_train_scaled = scaler.transform(X_oversampled)
X_valid_scaled = scaler.transform(X_valid)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.cluster import KMeans

cluster_model = KMeans(n_clusters=5)
cluster_model.fit(train_no_address)
cluster_pred = cluster_model.predict(X_valid_scaled)
matrix = confusion_matrix(y_valid, cluster_pred)
matrix
```

```python
score = f1_score(y_valid, cluster_pred, average='macro')
print('f1_Score is: ', score, ' for a kmeans clustering')
```

### PCA

```python
from sklearn.decomposition import PCA

pca = PCA(random_state=17, n_components=.95)
pca.fit(train_no_address)
X_train_pca = pca.transform(X_train_scaled)
X_valid_pca = pca.transform(X_valid_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
random_forest_balanced.fit(X_train_pca, y_oversampled)
pca_predictions = random_forest_balanced.predict(X_valid)
matrix = confusion_matrix(y_valid, pca_predictions)
matrix
```

```python
score = f1_score(y_valid, pca_predictions, average='macro')
print('f1_Score is: ', score, ' for a balanced random forest trained with PCA')
```

```python
print(classification_report(y_valid, pca_predictions, target_names=label_encoder.classes_))
```

### Neural Network attempt

```python
X_train_scaled = X_train_scaled.reshape(X_train_scaled.shape[0], 1, X_train_scaled.shape[1])
X_valid_scaled = X_valid_scaled.reshape(X_valid_scaled.shape[0], 1, X_valid_scaled.shape[1])
X_test_scaled = X_test_scaled.reshape(X_test_scaled.shape[0], 1, X_test_scaled.shape[1])
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import regularizers

model3 = Sequential()
model3.add(layers.SimpleRNN(units=100))
model3.add(layers.Dense(units=200))
model3.add(layers.Dense(units=5))
model3.compile(loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True), optimizer='adam', metrics=['sparse_categorical_accuracy'])
```

```python
history2 = model3.fit(
    X_train_scaled, y_oversampled,
    epochs=100, batch_size=25,
    validation_data=(X_valid_scaled, y_valid),
    verbose=2, shuffle=False
)
```

```python
plt.plot(history2.history['loss'], label='train')
plt.plot(history2.history['val_loss'], label='test')
plt.legend()
plt.show()
```
